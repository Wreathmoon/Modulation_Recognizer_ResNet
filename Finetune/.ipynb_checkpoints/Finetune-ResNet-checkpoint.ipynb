```python
import os
import gc
import json
import h5py
import keras
import random
import matplotlib
import numpy as np
import seaborn as sns
import scipy.io as sio
import scipy.io.wavfile
import tensorflow as tf
import matplotlib.pyplot as plt


from glob import glob
from keras import layers
from keras.models import Model
from keras.regularizers import *
from keras.optimizers import adam_v2
from keras.callbacks import TensorBoard, EarlyStopping
from keras import metrics, regularizers, optimizers, backend
from keras.layers import Reshape,ZeroPadding2D,Flatten,Activation
from keras.layers import Flatten,pooling,AlphaDropout,MaxPooling2D
from keras.layers import Input,Dense,Dropout,BatchNormalization,Conv2D

os.environ["KERAS_BACKEND"] = "tensorflow"
```

## 读数据

```python
#幅度缩小函数，使其与原模型数据近似
def shrink(data_big):
    data_temp = data_big.T
    data_small = np.zeros((2,1024,1))
    for i in range(data_temp.shape[0]):
        for j in range(data_temp.shape[1]):
            data_small[i][j] = float(data_temp[i][j]/1000)
    return data_small.T
```

```python
#读取real_dataset文件夹下所有文件夹名作为调制类型种类
filepath_datasets = '../Dataset/Real_dataset/'
classes = os.listdir(filepath_datasets)

#分调制类型读取信号数据
for modu_type in classes:
    print(modu_type)
    files= os.listdir(filepath_datasets+modu_type) 
    
    for file in files：        
        #用scipy.io.wavfile来读.wav文件，可忽略header
        rate, data_raw = scipy.io.wavfile.read(filepath_datasets+modu_type+'//'+file, mmap=False)
        data = np.expand_dims(data_raw,axis=0)  #把都进来的数据弄到和输入模型的数据一个维度的
        
        #每个.wav文件随机采1024个连续的点，采1000组
        for i in range(1000):
            slice = int(np.random.choice(range(0,data.shape[1]-1024), size=1, replace=False))
            data_slice = data[:,slice:(slice+1024),:]
            data_input = shrink(data_slice)  #缩小幅度
            
            label_raw = classes.index(modu_type) #标签
            y_init = np.zeros(24)
            y_init[label_raw] = 1
            y_init = np.expand_dims(y_init,axis=0)
            
            if modu_type==classes[0] and file == files[0] and i == 0:
                x = data_input
                y = y_init
            else:
                x = np.vstack((x,data_input))
                y = np.vstack((y,y_init))
```

```python
#保存了再读 下次就不用现生成数据集了
np.save(r'./npy/x.npy',x)
np.save(r'./npy/y.npy',y)
```

```python
x_load= np.load(r'./npy/x.npy')
y_load= np.load(r'./npy/y.npy')
```

```python
# 全部读取后再进行打乱index
data_idx = [i for i in range(x_load.shape[0])]
random.shuffle(data_idx)

x_data = x_load[data_idx]
y_data = y_load[data_idx]

# 划分训练集，验证集，测试集。8：1：1
total_num = x_data.shape[0]
train_num = int(total_num * 0.8)   # 先把训练集随机划出来
train_idx = np.random.choice(range(0,total_num), size=train_num, replace=False)
val_test_idx = list(set(range(0,total_num))-set(train_idx))  

x_train = x_data[train_idx]
y_train = y_data[train_idx]
x_val_test = x_data[val_test_idx]
y_val_test = y_data[val_test_idx]

val_test_num = x_val_test.shape[0]
val_num = int(val_test_num * 0.5)  # 再1：1随机划验证集和测试集
val_idx = np.random.choice(range(0,val_test_num), size=val_num, replace=False)
test_idx = list(set(range(0,val_test_num))-set(val_idx)) 

x_val = x_val_test[val_idx]
y_val = y_val_test[val_idx]
x_test = x_val_test[test_idx]
y_test = y_val_test[test_idx]

# 观测数据集的shape
print('train of x：',x_train.shape)
print('train of y：',y_train.shape)
print('val of x：',x_val.shape)
print('val of y：',y_val.shape)
print('test of x：',x_test.shape)
print('test of y：',y_test.shape)
```

## ResNet 模型

```python
#读取之前训练好的模型
filepath_weights =  '../Train/Weights/ResNet_weights_067-0.861_SNRover0.h5'
loaded_model = keras. models.load_model(filepath_weights)
```

```python
num_classes = len(classes)

#去掉原模型最后两层换成新的
x = loaded_model.layers[-3].output
#换成新的调制种类数要报错，就还用的之前的标签数组结构，之后多的值都设为0    
predictions = Dense(24, activation='softmax', name="dense_train")(x) 
    
finetune_model = Model(inputs=loaded_model.input, outputs=predictions)

adam = keras.optimizers.adam_v2.Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
finetune_model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['categorical_accuracy'])
finetune_model.summary()
```

## 训练

```python
print(tf.test.gpu_device_name()) # 看显卡是否调用成功
print('Start Traning:')
filepath = './weights.FinetuneResNet_weights_ep{epoch:03d}_{val_categorical_accuracy:.3f}.h5'

# 当连续 10 次迭代准确度不再增加时停止
earlystop = keras.callbacks.EarlyStopping(monitor='val_categorical_accuracy', patience=10, verbose=1, mode='auto')

# 保存准确度最高时的权值和模型
ckpt = keras.callbacks.ModelCheckpoint(filepath, monitor='val_categorical_accuracy',verbose=1, save_best_only=True, mode='auto')

finetune_train = finetune_model.fit(x_train,
                                    y_train,
                                    batch_size=1024,   # 不爆内存的最大batch size了
                                    epochs=1000,       
                                    verbose=1,
                                    validation_data=(x_val, y_val),
                                    callbacks = [ckpt,earlystop])  
```

## 画整个训练中损失函数和准确性的变化图

```python
plt.figure(figsize=(8,3)) 

# 保存损失和准确性值到txt
val_loss_list = finetune_train.history['val_loss']
loss_list = finetune_train.history['loss']
categorical_accuracy = finetune_train.history['categorical_accuracy']
val_categorical_accuracy = finetune_train.history['val_categorical_accuracy']

# 续写模式 不覆盖之前的数据
with open("./loss_acc/loss_acc_data.txt","w") as txt:
    txt.write('\n'+f'val_loss_list loss_list categorical_accuracy val_categorical_accuracy'+'\n')
    for i in range(len(loss_list)):
        txt.write(f'{val_loss_list[i]} {loss_list[i]} {categorical_accuracy[i]} {val_categorical_accuracy[i]}\n')
    txt.close()

# 并列着画
ax1=plt.subplot(121)
# 蓝色是训练集的 红色是验证集的
plt.plot(range(len(loss_list)),val_loss_list,color='red',label='val_loss')
plt.plot(range(len(loss_list)),loss_list,color='blue',label='train_loss')

ax2 = plt.subplot(122)
plt.plot(range(len(categorical_accuracy)),categorical_accuracy,color='blue',label='train_accuracy')
plt.plot(range(len(categorical_accuracy)),val_categorical_accuracy,color='red',label='val_accuracy')

plt.show()
```

## 加载表现最好的权重模型

```python
filepath_weights =  './weights/FinetuneResNet_weights_ep046_0.850.h5'
model = keras. models.load_model(filepath_weights)
```

## 测试集上进行测试

```python
filepath_datasets = '../Dataset/Real_dataset/'
classes = os.listdir(filepath_datasets)

# 预测值
y_test_pd = model.predict(x_test, batch_size=1024)

conf = np.zeros([len(classes),len(classes)])
confnorm = np.zeros([len(classes),len(classes)])

# j为实际调制类别，k为预测的
for i in range(0,x_test.shape[0]):
    j = list(y_test[i,:]).index(1)
    k = int(np.argmax(y_test_pd[i,:]))
    conf[j,k] = conf[j,k] + 1
    
# 归为0到1的概率
for i in range(0,len(classes)):
    confnorm[i,:] = conf[i,:] / np.sum(conf[i,:])
```

## 画各调制信号的分类准确度

```python
# jk对角线即为该类别的预测准确度
classes_pd = np.array([round(confnorm[i,i],4) for i in range(len(confnorm))])

temp1_classes_pd = classes_pd[:]
temp2_classes_pd = [int(temp1_classes_pd[i]*10000) for i in range(len(temp1_classes_pd))]

# 按准确度排序
sort_index = np.argsort(temp2_classes_pd)
classes = np.array(classes)
classes_sort = classes[sort_index]
classes_pd_sort = classes_pd[sort_index]

fig,ax = plt.subplots()
# 横着画柱状图
a = plt.barh(range(len(classes_sort)),classes_pd_sort,color='#6699CC')
for x in a:
    width = x.get_width()
    # 标注准确度百分比
    plt.text(width, x.get_y()+x.get_height()/2,'{:.2f}%'.format(width*100), ha='left',va='center',fontsize=8)
ax.set_yticks(range(len(classes_sort)))
ax.set_yticklabels(classes_sort)
plt.title('Categorical Accuracy of Classes', loc='center', fontsize='15',fontweight='bold')

plt.show()    
```
