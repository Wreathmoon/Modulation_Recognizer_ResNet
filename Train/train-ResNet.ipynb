```python
import os
import gc
import json
import h5py
import keras
import random
import matplotlib
import numpy as np
import seaborn as sns
import scipy.io as sio
import scipy.io.wavfile
import tensorflow as tf
import matplotlib.pyplot as plt

from keras import layers
from keras.models import Model
from keras.regularizers import *
from keras.optimizers import adam_v2
from keras.callbacks import TensorBoard, EarlyStopping
from keras import metrics, regularizers, optimizers, backend
from keras.layers import Reshape,ZeroPadding2D,Flatten,Activation
from keras.layers import Flatten,pooling,AlphaDropout,MaxPooling2D
from keras.layers import Input,Dense,Dropout,BatchNormalization,Conv2D

os.environ["KERAS_BACKEND"] = "tensorflow"
```

## 读数据

```python
'''由于内存有限，原始数据集从 2555904x1024x2 抽取到 374400x1024x2
（600*26*24，24种调制，26个信噪比，每个信号IQ两组数据）。
而且提取的数据集也被分成24个部分，分多次读取以节省内存'''

classes = ['32PSK', '16APSK', '32QAM', 'FM', 'GMSK', '32APSK', 'OQPSK',
           '8ASK', 'BPSK', '8PSK', 'AM-SSB-SC', '4ASK', '16PSK', '64APSK',
           '128QAM', '128APSK', 'AM-DSB-SC', 'AM-SSB-WC', '64QAM', 'QPSK',
           '256QAM', 'AM-DSB-WC', 'OOK','16QAM']

for i in range(0,24):  #分24次读取
    
    filename = '../Dataset/ExtractDataset/part'+str(i) + '.h5'
    f = h5py.File(filename,'r')
    
    # x为信号数据，y是标签（24种调制），z是信噪比标签
    x_temp = f["X"][:]
    y_temp = f["Y"][:]
    z_temp = f["Z"][:]
    f.close()       #读完就关，节省内存
    
    if i == 0:
        x_data = x_temp
        y_data = y_temp
        z_data = z_temp
    else:           #把之后读的数据拼起来
        x_data = np.vstack((x_data, x_temp))
        y_data = np.vstack((y_data, y_temp))
        z_data = np.vstack((z_data, z_temp))
```

## 打乱 & 划分数据集验证集测试集

```python
# 只抽取信噪比大于0的信号数据来进行训练
data_idx_allSNRs = [i for i in range(x_data.shape[0])]
data_idx = [i for i in data_idx_allSNRs if list(z_data[i])[0] >= 0] 

# 全部读取后再进行打乱index
random.shuffle(data_idx)

x_data = x_data[data_idx]
y_data = y_data[data_idx]
z_data = z_data[data_idx]

# 划分训练集，验证集，测试集。8：1：1
total_num = x_data.shape[0]
train_num = int(total_num * 0.8)   # 先把训练集随机划出来
train_idx = np.random.choice(range(0,total_num), size=train_num, replace=False)
val_test_idx = list(set(range(0,total_num))-set(train_idx))  

x_train = x_data[train_idx]
y_train = y_data[train_idx]
z_train = z_data[train_idx]
x_val_test = x_data[val_test_idx]
y_val_test = y_data[val_test_idx]
z_val_test = z_data[val_test_idx]

val_test_num = x_val_test.shape[0]
val_num = int(val_test_num * 0.5)  # 再1：1随机划验证集和测试集
val_idx = np.random.choice(range(0,val_test_num), size=val_num, replace=False)
test_idx = list(set(range(0,val_test_num))-set(val_idx)) 

x_val = x_val_test[val_idx]
y_val = y_val_test[val_idx]
z_val = z_val_test[val_idx]
x_test = x_val_test[test_idx]
y_test = y_val_test[test_idx]
z_test = z_val_test[test_idx]

# 观测数据集的shape
print('train of x：',x_train.shape)
print('train of y：',y_train.shape)
print('train of z：',z_train.shape)
print('val of x：',x_val.shape)
print('val of y：',y_val.shape)
print('val of z：',z_val.shape)
print('test of x：',x_test.shape)
print('test of y：',y_test.shape)
print('test of z：',z_test.shape)
```

## Resnet 模块

```python
def residual_stack(X, kennel_size, Seq, pool_size):
    
    # 1*1 线性 卷积
    X = Conv2D(32, (1, 1), padding='same', name=Seq+"_conv1", kernel_initializer='glorot_normal',data_format='channels_first')(X)
    
    # Residual 单元
    X_shortcut = X
    X = Conv2D(32, kennel_size, padding='same',activation="relu",name=Seq+"_conv2", kernel_initializer='glorot_normal',data_format='channels_first')(X)
    X = Conv2D(32, kennel_size, padding='same', name=Seq+"_conv3", kernel_initializer='glorot_normal',data_format='channels_first')(X)
    X = layers.add([X,X_shortcut]) # 短路
    X = Activation("relu")(X)
    
    # Residual 单元
    X_shortcut = X
    X = Conv2D(32, kennel_size, padding='same',activation="relu",name=Seq+"_conv4", kernel_initializer='glorot_normal',data_format='channels_first')(X)
    X = Conv2D(32, kennel_size, padding='same', name=Seq+"_conv5", kernel_initializer='glorot_normal',data_format='channels_first')(X)
    X = layers.add([X,X_shortcut]) # 短路
    X = Activation("relu")(X)
    
    # MaxPooling
    X = MaxPooling2D(pool_size=pool_size, strides=pool_size, padding='valid', data_format='channels_first')(X)
    
    return X
```

## ResNet 模型

```python
# 输入层
X_input = Input(x_train.shape[1:])
X = Reshape([1,1024,2], input_shape=x_train.shape[1:])(X_input)

# Residual Stack
X = residual_stack(X,kennel_size=(3,2),Seq="ReStk0",pool_size=(2,2))   #shape:(512,1,32)
X = residual_stack(X,kennel_size=(3,1),Seq="ReStk1",pool_size=(2,1))   #shape:(256,1,32)
X = residual_stack(X,kennel_size=(3,1),Seq="ReStk2",pool_size=(2,1))   #shape:(128,1,32)
X = residual_stack(X,kennel_size=(3,1),Seq="ReStk3",pool_size=(2,1))   #shape:(64,1,32)
X = residual_stack(X,kennel_size=(3,1),Seq="ReStk4",pool_size=(2,1))   #shape:(32,1,32)
X = residual_stack(X,kennel_size=(3,1),Seq="ReStk5",pool_size=(2,1))   #shape:(16,1,32)

# 全连接层
X = Flatten(data_format='channels_first')(X)
X = Dense(128, activation='selu', kernel_initializer='glorot_normal', name="dense1")(X)
X = AlphaDropout(0.3)(X)  #0.3 better than 0.5
X = Dense(len(classes), kernel_initializer='glorot_normal', name="dense2")(X)

# 激活层 分类 SoftMax
X = Activation('softmax')(X)

# 创建模型
def create_model():
    from keras.models import Model
    model = Model(inputs=X_input,outputs=X)    
    return model

adam = keras.optimizers.adam_v2.Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
resnet_model = create_model()
resnet_model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['categorical_accuracy'])
resnet_model.summary()
```

## 训练

```python
print(tf.test.gpu_device_name()) # 看显卡
print('Start Traning:')

#保存模型路径
filepath_weights = r'./Weights/ResNet_weights_ep{epoch:03d}_{val_categorical_accuracy:.3f}.h5'

# 当连续 10 次迭代准确度不再增加时停止
earlystop = keras.callbacks.EarlyStopping(monitor='val_categorical_accuracy', patience=10, verbose=1, mode='auto')

# 保存准确度最高时的权值和模型
ckpt = keras.callbacks.ModelCheckpoint(filepath_weights, monitor='val_categorical_accuracy',verbose=1, save_best_only=True, mode='auto')

# 从之前的检查点加载并继续训练
# resnet_model = keras.models.load_model(filepath_weights) 

resnet_train = resnet_model.fit(x_train,
                                y_train,
                                batch_size=256,   # 不爆内存的最大batch size了
                                epochs=100,       # 一般跑不完就收敛了
                                verbose=1,
                                validation_data=(x_val, y_val),
                                callbacks = [ckpt,earlystop])  
```

### 拉伸函数 & 缩小函数

```python
#拉伸函数
def stretch(data_input):
    data_input_T = data_input[:,:int(data_input.shape[1]/2),:].T
    data_output = np.zeros((2,2*data_input_T.shape[1],1))
    for i in range(data_output.shape[1]):
        if i%2 == 0:
            data_output[0][i] = data_input_T[0][int(i/2)]
            data_output[1][i] = data_input_T[1][int(i/2)]
        elif (i == data_output.shape[1]-1):
            data_output[0][i] = data_input_T[0][data_input_T.shape[1]-1]
            data_output[1][i] = data_input_T[1][data_input_T.shape[1]-1]
        else:
            data_output[0][i] = (data_input_T[0][int((i+1)/2)]+data_input_T[0][int((i-1)/2)])/2
            data_output[1][i] = (data_input_T[1][int((i+1)/2)]+data_input_T[1][int((i-1)/2)])/2
    data_output = data_output.T
    return data_output

#多次拉伸
def multi_stretch(n,data_i):
    if n == 0:
        return data_i
    return multi_stretch(n-1,stretch(data_i))

#缩小信号幅度的函数
def shrink(data_big):
    data_temp = data_big.T
    data_small = np.zeros((2,1024,1))
    for i in range(data_temp.shape[0]):
        for j in range(data_temp.shape[1]):
            data_small[i][j] = float(data_temp[i][j]/1000)
    return data_small.T
```
